```python
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
##Suppose your clients have changed their minds about the labels you created in your â€œbusiest_dayâ€ column. Now, they want â€œBusiest dayâ€ to become â€œBusiest daysâ€ (plural). 
result2 = []
for value in df["order_dow"]:
  if (value == 0) | (value == 1):
    result2.append("Busiest days")
  elif (value == 3) | (value == 4):
    result2.append("Slowest days")
  else:
    result2.append("Regularly days")
```

```python
##When too many users make Instacart orders at the same time, the app freezes.
hours = []
for value in df["order_hour_of_day"]:
  if (value >= 10) & (value <= 15):
    hours.append("Most orders")
  elif (value >= 0) & (value <= 6) | (value == 23):
    hours.append("Fewest orders")
  else:
    hours.append("Average orders")
```

```python
hours
```

```python
##â€œMost orders,â€ â€œAverage orders,â€ and â€œFewest orders.â€ Create a new column containing these labels called â€œbusiest_period_of_day.â€
df["busiest_period_of_day"] = hours
```
