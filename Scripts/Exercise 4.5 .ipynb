```python
#creating subset to view the missing data in specific column
df_nan = df_prods[df_prods['product_name'].isnull() == True]
df_nan
```

```python
#creating data frame that has no missing data
df_prods_clean = df_prods[df_prods['product_name'].isnull() == False]
df_prods_clean
```

```python
#The following command will look for full duplicates within your dataframe:

df_dups = df_prods_clean[df_prods_clean.duplicated()]
df_dups
```

```python
#create a new dataframe that doesn’t include the duplicates you just identified using the drop_duplicates() function:

df_prods_clean_no_dups = df_prods_clean.drop_duplicates()
df_prods_clean_no_dups.shape
```

```python
#exporting data
df_prods_clean_no_dups.to_csv(os.path.join(path, 'Data','Prepared Data', 'clean_products.csv'))
```
