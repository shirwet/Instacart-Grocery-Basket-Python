```python
##creating loyality if statement  based on max_order
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
##Q5: The marketing team at Instacart wants to know whether thereâ€™s a difference between the spending habits of the three types of customers you identified.
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max', 'sum']})
```

```python
##it shows that loyal customers are purchasing cheaper products than new and regualr, also it shows that new customer average prices are more than others spend
```

```python
##The team now wants to target different types of spenders in their marketing campaigns
ords_prods_merge['avg_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
##Create a spending flag for each user based on the average price across all their orders using the following criteri
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_price'] > 10, 'spending_flag'] = 'High spender'
ords_prods_merge.loc[ords_prods_merge['avg_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
##the Instacart team wants you to determine frequent versus non-frequent customers.
ords_prods_merge['mdian_freq'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
##
ords_prods_merge.loc[ords_prods_merge['mdian_freq'] > 20, 'Cust_frequency'] = 'Non-frequent customer'
ords_prods_merge.loc[ords_prods_merge['mdian_freq'] > 10 & (ords_prods_merge['mdian_freq'] <= 20), 'Cust_frequency'] = 'Non-frequent customer'
ords_prods_merge.loc[ords_prods_merge['mdian_freq'] <= 10, 'Cust_frequency'] = 'Frequent customer'
```
